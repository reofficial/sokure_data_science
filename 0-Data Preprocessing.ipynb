```python
import pandas as pd

df = pd.read_pickle('sokureplays.pkl')
```

```python
# Convert the Unix timestamp to datetime
df['datetime'] = pd.to_datetime(df['time'], unit='ns')

# Extract date, month, and year from the datetime column
df['date'] = df['datetime'].dt.day
df['month'] = df['datetime'].dt.month
df['year'] = df['datetime'].dt.year
df.drop(columns=['time'], inplace=True)
```

```python
df_server = df[['server_player_id','server_rounds', 'server_character', 'client_character', 'server_cards', 'server_nick', 'server_country', 'datetime', 'date', 'month', 'year']].copy()
df_client = df[['client_player_id','client_rounds', 'client_character', 'server_character', 'client_cards', 'client_nick', 'client_country', 'datetime', 'date', 'month', 'year']].copy()

df_server.rename(columns={'server_player_id': 'player_id', 'server_rounds': 'rounds', 'server_character': 'character', 'client_character': 'opponent', 'server_cards': 'cards', 'server_nick': 'nick', 'server_country': 'country'}, inplace=True)
df_client.rename(columns={'client_player_id': 'player_id', 'client_rounds': 'rounds', 'client_character': 'character', 'server_character': 'opponent', 'client_cards': 'cards', 'client_nick': 'nick', 'client_country': 'country'}, inplace=True)

df_server['is_server'] = 1
df_client['is_server'] = 0

join = pd.concat([df_server, df_client], ignore_index=True)

df = join
```

```python
from pycountry_convert import country_alpha2_to_country_name, country_name_to_country_alpha3

# Define a function to convert alpha-2 code to country name and then to alpha-3 code
def convert_country_code(alpha2_code):
    if alpha2_code.strip():  # Check if alpha2_code is not empty or whitespace
        try:
            country_name = country_alpha2_to_country_name(alpha2_code.upper(), cn_name_format="default")
            alpha3_code = country_name_to_country_alpha3(country_name, cn_name_format="default")
            return alpha3_code
        except KeyError:
            return None  # Return None if country code is invalid or not found
    else:
        return None  # Return None if alpha2_code is empty or whitespace

# Apply the conversion function to each row in 'country' column
df['country'] = df['country'].apply(convert_country_code)
```
